# Step 6

```python
# Creating a histogram to see distribution
sns.histplot(df['Happiness Score'], bins = 10, kde = True)
```

```python
# Creating score category for categorical plot
df.loc[df['Happiness Score'] < 4.5 , 'Score category'] = 'Low'
df.loc[(df['Happiness Score'] >= 4.5) & (df['Happiness Score'] < 6.5), 'Score category'] = 'Middle'
df.loc[df['Happiness Score'] >= 6.5, 'Score category'] = 'High'
```

```python
df['Score category'].value_counts(dropna = False)
```

```python
# Creating categorical plot with Economy (GDP per Capita)
sns.set(style="ticks")
cat_plot_GDP = sns.catplot(x='Economy (GDP per Capita)', y='Happiness Score', hue="Score category", data=df)
```

```python
# Creating categorical plot with Health (Life Expectancy)
sns.set(style="ticks")
cat_plot_life = sns.catplot(x='Health (Life Expectancy)', y='Happiness Score', hue="Score category", data=df)
```

```python
# Creating categorical plot with Trust (Government Corruption)
sns.set(style="ticks")
cat_plot_life = sns.catplot(x='Trust (Government Corruption)', y='Happiness Score', hue="Score category", data=df)
```

```python
# Creating categorical plot with Freedom
sns.set(style="ticks")
cat_plot_life = sns.catplot(x='Freedom', y='Happiness Score', hue="Score category", data=df)
```

#### There is no categorical plot that does not reveal a clear distinction between low, moderate, and high happiness scores. And also it does not appear to be any significant outliers in this data set across variables.
